```python
"""
This script ingest data from PDF files and Youtube channels. 
"""

from embedchain import App
from dotenv import load_dotenv

load_dotenv()
```

```python
app = App.from_config(config_path="ingestion/youtube_video.yaml")

urls = ["https://www.youtube.com/watch?v=OFkHI7b9aWQ&list=PLsmVMagkebHkYLmXfT1quCNLy4gy48xdH&index=1&pp=iAQB",
        "https://www.youtube.com/watch?v=xOpCmCM84_c&list=PLsmVMagkebHkYLmXfT1quCNLy4gy48xdH&index=2&pp=iAQB",
        "https://www.youtube.com/watch?v=eSgGucsShOA&list=PLsmVMagkebHkYLmXfT1quCNLy4gy48xdH&index=3&pp=iAQB"]
for url in urls:
    app.add(url, data_type="youtube_video")
```

```python
answer, sources = app.query("Resuma os vídeos, trazendo os significados de cada um dos números apresentados nos vídeos.", citations=True)
print(answer)
# Answer: The net worth of Elon Musk is $221.9 billion.

```
